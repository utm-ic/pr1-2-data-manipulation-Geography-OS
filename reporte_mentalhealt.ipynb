### Detección de atípicos, también conocidos cómo Outliers, utilizaré dos métodos para su detección
a) Método por desviación estándar
b) Método cuartílico

Ya realicé el intento de calcular los estadisticos básicos, para detectar atípicos, sin embargo, tenemos datos nulos y variables de texto que impiden los cálculos, realizaré:

    a) .Describe, para determinar alguna medidas en el set de datos origen
    b) Tomaré unicamente las columnas numéricas
    c) Estoy decidiendo si hago una imputación de datos por la media, el método clasico, porque si elimino los nulos será la mayoría de los datos, según mis cálculos realizados antes.

```python
#Para resolver el inciso a) realizando el .describe, funciona
Df_MentalHealt.describe()
```

```python
#Esta operación me permite filtrar los nombres de las columnas son numéricas
col_numint = MentalhType[MentalhType =='int64'].index
col_numint
```

```python
# Sólo consideraré las columnas numéricas, despues de haber obtenido los nombres de
#de las columnas que son numéricas, generaré una lista, par después 
#comparar esos nombres con los del set de datos original, 
#utilizaré numpy para obtener los valores absolutos
import numpy as np
ColsNumMentalH = [
    'Age',
    'Social_Support_Score',
    'Anxiety_Score',
    'Depression_Score',
    'Stress_Level',
    'Family_History_Mental_Illness',
    'Chronic_Illnesses',
    'Therapy',
    'Meditation',
    'Financial_Stress',
    'Work_Stress',
    'Self_Esteem_Score',
    'Life_Satisfaction_Score',
    'Loneliness_Score'
    ]

meanMentalH = Df_MentalHealt[ColsNumMentalH].mean()
stdMentalH = Df_MentalHealt[ColsNumMentalH].std()

#Vamos a asegurarnos de que todas nuestra columnas en efcto sean numéricas
for column in ColsNumMentalH:
    Df_MentalHealt[column] = pd.to_numeric(Df_MentalHealt[column], errors='coerce')


Umbral = 3 #Todos los datos que rebasen el calculo de las desviaciones
oulierMentalH = Df_MentalHealt[(np.abs(Df_MentalHealt[ColsNumMentalH] - meanMentalH)>Umbral*stdMentalH).any(axis=1)]
print('Media de las variables ','\n',meanMentalH,end='\n')
print('Mediana de las variables ','\n',stdMentalH,end='\n')
print('Datos atípicos encontrados')
print(oulierMentalH)
```

Esto es frustrante, no hay atípicos por el método de media!!

```python
# Detección de outliers por el método intercuartilico, lo haremos sobre las columnas numéricas
ColsNumMentalH = [
    'Age',
    'Social_Support_Score',
    'Anxiety_Score',
    'Depression_Score',
    'Stress_Level',
    'Family_History_Mental_Illness',
    'Chronic_Illnesses',
    'Therapy',
    'Meditation',
    'Financial_Stress',
    'Work_Stress',
    'Self_Esteem_Score',
    'Life_Satisfaction_Score',
    'Loneliness_Score'
    ]
Q1 = Df_MentalHealt[ColsNumMentalH].quantile(0.25)
Q3 = Df_MentalHealt[ColsNumMentalH].quantile(0.75)
IQR = Q3 - Q1
condition = ((Df_MentalHealt[ColsNumMentalH] < (Q1 - 1.5 * IQR)) | (Df_MentalHealt[ColsNumMentalH] > (Q3 + 1.5 * IQR))).any(axis=1)
outliers = Df_MentalHealt[condition]
print("Datos atípicos encontrados, para las columnas numéricas!")
print(outliers)
```

####  Seleccionamos las filas que contienen datos atípicos en las columnas numéricas especificadas. Estas filas son aquellas donde al menos una de las columnas numéricas tiene un valor que está fuera del rango definido por el método de los cuartiles (más de 1.5 veces el rango intercuartílico por debajo del primer cuartil o por encima del tercer cuartil)
Haré una pequeña gráfica para apreciar su dispersión

```python
import matplotlib.pyplot as plt

#Crearemos la figura base
fig, ax = plt.subplots(figsize=(12,8))

#Grafica de los datos atípicos
for column in ColsNumMentalH:
    ax.scatter(outliers.index, outliers[column], label=column)
    
#Añadiendo etiquetas
ax.set_xlabel('Indice')
ax.set_ylabel('Valor')
ax.set_title('Datos atipicos detectados')    
```

Realizaremos una imputación de datos, por un método sencillo, con la media de cada columna, ya que si eliminamos los registros con datos nulos perderemos gran parte de la información, 1040 registros.

```python
#Vamos a llenar los datos nulos, primero de las varaibles numéricas, despues haremos por las categorícas
ColsNumMentalH = [
    'Age',
    'Social_Support_Score',
    'Anxiety_Score',
    'Depression_Score',
    'Stress_Level',
    'Family_History_Mental_Illness',
    'Chronic_Illnesses',
    'Therapy',
    'Meditation',
    'Financial_Stress',
    'Work_Stress',
    'Self_Esteem_Score',
    'Life_Satisfaction_Score',
    'Loneliness_Score'
    ]
Df_MentalHealt[ColsNumMentalH] = Df_MentalHealt[ColsNumMentalH].fillna(Df_MentalHealt[ColsNumMentalH].median())
Df_MentalHealt.info()
```

```python
#Despues de la imputación a las variables numéricas, trabajermos las categóricas, asignaremos el valor maás frecuente a los nulos
col_numint = MentalhType[MentalhType =='object'].index
col_numint
```

```python
#Afectar las variables categóricas
# Seleccionar todas las columnas categóricas automáticamente
ColsCatMentalH = Df_MentalHealt.select_dtypes(include=['object', 'category']).columns

# Imputar valores nulos con la moda de cada columna
for col in ColsCatMentalH:
    Df_MentalHealt.fillna({col:Df_MentalHealt[col].mode()[0]},inplace=True)
Df_MentalHealt.info()    
```
