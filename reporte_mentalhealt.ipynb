#### C 
Las variables se suelen clasificar para un mejor análisis de datos, las de *texto* se suelen llamar **categóricas** o **nominales**. 
Estas variables representan categorías o grupos y no tienen un orden intrínseco. Ejemplos en nuestro set de datos: Education_Level, Substance_Use, Medication_Use, Gender, etc.

Las variables numéricas se dividen en dos tipos principales:

1. **Variables continuas**: Pueden tomar cualquier valor dentro de un rango. Ejemplos en nuestro set de datos: Sleep_Hours, Physical_Activity_Hrs, Age, etc.
2. **Variables discretas**: Solo pueden tomar valores específicos y contables. Ejemplos: Social_Support_Score, Anxiety_Score, Chronic_Illnesses, etc.

Ambos tipos de variables *numéricas* se agrupan bajo el término **variables cuantitativas**.
