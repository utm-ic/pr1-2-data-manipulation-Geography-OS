## Detectando valores nulos y no nulos
Utilizaremos la función .isnull(), con el fin de detectar los valores nulos en el dataframe, se rezalizarán dos análisis, por filas y el conteo total de valores.

```python
'''
Para calcular el porcentaje de filas con valores nulos en el dataset no es suficiente con sumar los 
totales obtenidos de las variables, pues incluso estará contando duplicados, debemos preveer que obtengamos
las filas que contienen nulos, para no duplicar.
'''
NulRowMentalH = (Df_MentalHealt.isnull().any(axis=1).sum() / len(Df_MentalHealt)) * 100
print(f"Pocentaje de filas con valores nulos: {NulRowMentalH:.2f}%")
```

```python
porcentaje_no_nulos = (Df_MentalHealt.notnull().sum().sum() / Df_MentalHealt.size) * 100
print(f"Porcentaje de valores no nulos: {porcentaje_no_nulos:.2f}%")
```

## En este ejercicio descubrí qué :
Los valores no nulos y las filas con al menos un nulo son métricas diferentes, por eso los números no necesariamente suman 100%., nuestro caso, 88.3 + 93.73 no es igual a 100%.
Aclaración de conceptos.

El Siguiente código obtiene en una misma escala de medición los porcentajes de nulos y no nulos, veamos utilizando una misma metrica, es cierto que obteniendo un porcentage el resto se puede deducir, pero no fue mi caso y realicé esta investigación. 

```python
total_nulos = Df_MentalHealt.isnull().sum().sum() 
total_no_nulos = Df_MentalHealt.notnull().sum().sum()

porcentaje_nulos = (Df_MentalHealt.isnull().sum().sum() / Df_MentalHealt.size) * 100
porcentaje_no_nulos = (Df_MentalHealt.notnull().sum().sum() / Df_MentalHealt.size) * 100

print(f"Total de valores nulos: {total_nulos:.2f}, Porcentaje de valores nulos: {porcentaje_nulos:.2f}%")
print(f"Total de valores no nulos: {total_no_nulos:.2f}, Porcentaje de valores no nulos: {porcentaje_no_nulos:.2f}%")
```

### Detectaré duplicados después de haber expulsado las variables con nulos.

### Dentro de este punto haré una detección de atípicos, también conocidos cómo Outliers, utilizaré dos métodos para su detección.
Primero eliminaré las variables que contienen los nulos
a) Método por desviación estándar
b) Método cuartílico

Ya realicé el intento de calcular los estadisticos básicos, para detectar atípicos, sin embargo, tenemos datos nulos y variables de texto que impiden los cálculos, realizaré:

    a) .Describe, para determinar alguna medidas en el set de datos origen
    b) Tomaré unicamente las columnas numéricas
    c) Estoy decidiendo si hago una imputación de datos por la media, el método clasico, porque si elimino los nulos será la mayoría de los datos, según mis cálculos realizados antes.

```python
# Detección de outliers por el método intercuartilico, lo haremos sobre las columnas numéricas
ColsNumMentalH = Df_MentalHealt.select_dtypes(include=['int64', 'float64']).columns
Q1 = Df_MentalHealt_clean[ColsNumMentalH].quantile(0.25)
Q3 = Df_MentalHealt_clean[ColsNumMentalH].quantile(0.75)
IQR = Q3 - Q1
condition = ((Df_MentalHealt_clean[ColsNumMentalH] < (Q1 - 1.5 * IQR)) | (Df_MentalHealt_clean[ColsNumMentalH] > (Q3 + 1.5 * IQR))).any(axis=1)
outlMentalH = Df_MentalHealt_clean[condition]
print("Datos atípicos encontrados, para las columnas numéricas!")
print(outlMentalH)
```

```python
import matplotlib.pyplot as plt

#Crearemos la figura base
fig, ax = plt.subplots(figsize=(12,8))

#Grafica de los datos atípicos
for column in ColsNumMentalH:
    ax.scatter(outlMentalH.index, outlMentalH[column], label=column)
    
#Añadiendo etiquetas
ax.set_xlabel('Indice')
ax.set_ylabel('Valor')
ax.set_title('Datos atipicos detectados')    
```

```python
# Filtrar personas con un puntaje alto de ansiedad
HanxietyMental = Df_MentalHealt_clean[Df_MentalHealt_clean['Anxiety_Score'] > 16]

# Calcular el porcentaje de personas con alta ansiedad
perHanxietyMental = (len(HanxietyMental) / len(Df_MentalHealt_clean)) * 100

print(f"Porcentaje de personas con nivel alto de ansiedad (>16): {perHanxietyMental:.2f}%")
```

¿Qué significa bins en este contexto?
En estadística y visualización de datos, un bin es un intervalo de valores dentro del cual agrupamos los datos. Se usa comúnmente para clasificar o agrupar valores continuos en categorías.

```python
#Comparación entre nivel de ansiedad y horas de sueño
# Crear categorías de horas de sueño, utilizaré en este ejercicio un concepto nuevo, de estadística, se llama ***bins***
bins = [0, 5, 7, float('inf')]
labels = ['Menos de 5h', 'Entre 5 y 7h', 'Más de 7h']

Df_MentalHealt_clean['Sleep_Category'] = pd.cut(Df_MentalHealt_clean['Sleep_Hours'], bins=bins, labels=labels)

# Calcular el nivel de estrés promedio en cada categoría
stress_by_sleep = Df_MentalHealt_clean.groupby('Sleep_Category', observed=True)['Stress_Level'].mean() #Para futuras actualizaciones se pondrá la variable observed en true, así que habrá que ponerlo de una vez

print("\nNivel promedio de estrés según las horas de sueño: ")
print(stress_by_sleep)
```

```python
#Realizaremos una visualización del caso para confirmar o descartar la hipótesis de que a myor horas de sueño mayor el nivel de estress
# Graficar el nivel de estrés promedio por grupo de sueño
x=6 # Esta varible nos servira para determinar la propiedad HUE del color en la paleta, se solicitará pra nuevas versiones
plt.figure(figsize=(8,5))
sns.barplot(x=stress_by_sleep.index, y=stress_by_sleep.values, palette= "coolwarm",legend=False, hue=x)
plt.title("Nivel de Estrés Promedio por Categoría de Sueño")
plt.xlabel("Horas de Sueño")
plt.ylabel("Nivel de Estrés Promedio")
plt.show()
```
